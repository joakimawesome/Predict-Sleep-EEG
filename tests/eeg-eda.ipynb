```python
import os
import numpy as np
import matplotlib.pyplot as plt
import mne
import seaborn as sns
```

```python
main_dir = os.path.dirname(os.getcwd())
file_EEG = os.listdir(main_dir + r'/data/EEG')
file_vhdr = [f for f in file_EEG if f.split('.')[-1] == 'vhdr']

idx = 0 # the experiment number
path_vhdr = "../data/EEG/" + file_vhdr[idx]

raw = mne.io.read_raw_brainvision(path_vhdr, preload=True)
raw
```

```python
raw.plot(duration=5, n_channels=len(raw.info['ch_names']));
```

```python
raw.plot_psd(average=True, verbose=False);
raw.notch_filter(60, verbose=False).plot_psd(average=True, verbose=False);
```

```python
signal = raw.copy()
signal.load_data()
```

```python
signal.plot(duration=10, start=1740, n_channels=2, highpass=0.1, lowpass=200.0, verbose=False);
```
